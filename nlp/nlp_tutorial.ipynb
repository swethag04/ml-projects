```python
import nltk
import pandas as pd
```

```python
from nltk import word_tokenize, sent_tokenize
nltk.download('punkt')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# word tokenize to split the text into individual elements of text
words = word_tokenize(principia)
print(words[:5])
```

```python
# sentence tokenize to split text int sentencens
sent = sent_tokenize(principia)
print(sent[:2])
```

```python
# create a collection of unique words
unique_words = set(word_tokenize(principia))
print(unique_words)
```

```python
# count of words
length = int(len(word_tokenize(principia)))
print(length)
```

```python
# Lexical diversity of a text is the ratio of unique to total words
words = len(word_tokenize(principia))
unique = len(set(word_tokenize(principia)))
lexical_diversity = float(unique/words)
print("Lexical diversity: ", lexical_diversity)
```

```python
# parts of speech
nltk.download('averaged_perceptron_tagger')
words_pos = nltk.pos_tag(word_tokenize(principia))
print(words_pos[:5])
```

```python
for word in nltk.ne_chunk(words_pos):
  if hasattr(word, 'label'):
    print(word)
    print(word.leaves())
    print(word.label())
```

```python
# creating a list of tuples (word, entity_type)
named_entities =[]
for word in nltk.ne_chunk(words_pos):
  if hasattr(word, 'label'):
    named_entities.append((' '.join(c[0] for c in word.leaves()), word.label()))
print(named_entities[:5])
```

```python
# Include only entities labeled 'GPE'
places = [i[0].lower() for i in named_entities if i[1] == 'GPE']
print(places[:5])
```

Feature extraction techniques to convert text into numerical feature matrices or vectors.
1. Bag of words
2. TF-IDF
3. Full word vectorization

Bag of words
* Represents text as a bag of words
* Create a feature for every distinct token in the data and enter the number of occurences of each token in that document
* Discards the ordering of words
* Does not track how informative words are


TF-IDF ( Term Frequency Inverse Document Frequency)
* TF-IDF score computed for every token, which is the product of TF and IDF
* TF = # times that `t` occurs in `d` / # words in `d`
* TF is the normalized version of bag of words
* IDF = -log(# documents that contain `t`/total # of documents)
* IDF amplifies rare words and annihilates words present in every document
* Indicates the importance of a particular word to a document in a collection

```python
sad_df = pd.read_csv('/content/sample_data/Emotion(sad).csv.zip', compression='zip')
```

```python
full_df = pd.concat([happy_df, sad_df]).reset_index(drop=True)
```

```python
X = full_df.drop('sentiment', axis=1)
y = full_df['sentiment']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
X_train, X_test, y_train, y_test = train_test_split(X['content'], y, random_state=42)
```

```python
X_train.head()
```

```python
# Create a bag of wrods representation of text
cvect = CountVectorizer()
dtm = cvect.fit_transform(X_train)
pd.DataFrame(dtm.toarray(), columns = cvect.get_feature_names_out()).head()
```

```python
# removing stopwords from text before vectorizing
cvect2 = CountVectorizer(stop_words='english')
X_train_vect_2 = cvect2.fit_transform(X_train)
X_test_vect_2 = cvect2.transform(X_test)
print(X_train_vect_2)
```

```python
# limiting features to top 300 words
cvect3 = CountVectorizer(stop_words='english', max_features=300)
X_train_vect_3 = cvect3.fit_transform(X_train)
X_test_vect_3 = cvect3.transform(X_test)
X_train_vect_3
```

```python
# Logistic Regression
vect_pipe_1 = Pipeline([('cvect', CountVectorizer()),
                        ('lgr', LogisticRegression())])
vect_pipe_1.fit(X_train, y_train)
print(vect_pipe_1.named_steps)
```

```python
test_acc = vect_pipe_1.score(vect_pipe_1.predict(X_test), y_test)
print(test_acc)
```

```python
params = {'cvect__max_features': [100, 500, 1000, 2000],
            'cvect__stop_words': ['english', None]}
```

```python
grid = GridSearchCV(param_grid = params,
                    estimator = vect_pipe_1)
grid.fit(X_train, y_train)
test_acc = grid.score(grid.predict(X_test), y_test)
print(grid.best_params_)
```
