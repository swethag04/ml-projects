```python
from statsmodels.tsa.seasonal import seasonal_decompose

import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sunspots = sm.datasets.sunspots.load_pandas()
air = pd.read_csv('sample_data/AirPassengers.csv')
```

```python
sunspots_df = sunspots.data
sunspots_df.head(5)
```

```python
sunspots_df.dtypes
```

```python
sunspots_df.index = pd.date_range(start='1700-01-01',
                                 periods= len(sunspots_df), freq='A')
sunspots_df.head()
```

```python
sunspots_df = sunspots_df.drop(columns=['YEAR'])
sunspots_df.head()
```

```python
sunspots_df.plot(figsize=(16,4))
```

```python
decompose = seasonal_decompose(sunspots_df, model='additive', period=11)
decompose.plot()
plt.show()
```
