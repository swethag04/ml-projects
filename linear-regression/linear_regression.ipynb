###Linear Regression
Models a linear relationship between independent and dependent variables

**Loss function**
*   Computes the badness of a prediction; a model with higher loss is considered worse
*   Characterizes the error that results from a choice of model or model parameters


**Common loss functions:**
1. L2 loss
*   Squared difference between prediction and actual (y-y')**2
*   Sensitive to outliers as it squares the difference
* L2 is a great choice to penalise large errors and outliers


2. L1 loss

*   Absolute difference between actual and prediction Abs(y-y')


**For non-numeric features**


*   Use one hot encoding





**Dataset and Objective**:
Explore the Diamond dataset from Seaborn and build a linear regression model to predict the price of diamond

```python
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression, HuberRegressor
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds['cut'].value_counts()
```

```python
diamonds['clarity'].value_counts()
```

```python
diamonds['color'].value_counts()
```

```python
# Build a linear regrssion model using all features
X = pd.get_dummies(diamonds.drop('price', axis = 1))
y= diamonds['price']
```

```python
X.head()
```

```python
linreg = LinearRegression(fit_intercept=False).fit(X,y)
pred = linreg.predict(X)
linreg_mse = mean_squared_error(pred, y )
print(linreg)
print(linreg.coef_)
print(linreg_mse)
```
