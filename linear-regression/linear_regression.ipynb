###Linear Regression
Models a linear relationship between independent and dependent variables

**Loss function**
*   Computes the badness of a prediction; a model with higher loss is considered worse
*   Characterizes the error that results from a choice of model or model parameters


**Common loss functions:**
1. L2 loss
*   Squared difference between prediction and actual (y-y')**2
*   Sensitive to outliers as it squares the difference
* L2 is a great choice to penalise large errors and outliers


2. L1 loss

*   Absolute difference between actual and prediction Abs(y-y')


**For non-numeric features**


*   Use one hot encoding





###Simple Linear Regression Model
**Dataset and Objective**:
Explore the Diamond dataset from Seaborn and build a linear regression model to predict the price of diamond

```python
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression, HuberRegressor
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds['cut'].value_counts()
```

```python
diamonds['clarity'].value_counts()
```

```python
diamonds['color'].value_counts()
```

```python
# Build a linear regrssion model using all features
X = pd.get_dummies(diamonds.drop('price', axis = 1))
y= diamonds['price']
```

```python
X.head()
```

```python
linreg = LinearRegression(fit_intercept=False).fit(X,y)
pred = linreg.predict(X)
linreg_mse = mean_squared_error(pred, y )
print(linreg)
print(linreg.coef_)
print(linreg_mse)
```

### Using Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
auto = pd.read_csv('sample_data/auto.csv')
auto.head()
```

Fitting a quadratic model using only the `horsepower` feature to predict `mpg`

```python
# Quad model
pipe = Pipeline([('quad_features', PolynomialFeatures(degree=2)),
                ('quad_model', LinearRegression())])
print(type(pipe))
print(pipe.named_steps)
```

```python
pipe.fit(auto[['horsepower']], auto['mpg'])
quad_pipe_mse= float(mean_squared_error(pipe.predict(auto[['horsepower']]), auto['mpg']))
print(quad_pipe_mse)
```

```python
quad_reg = pipe.named_steps['quad_model']
coefs = quad_reg.coef_
print(coefs)
```

The coeffecients have 3 values. The model contains an intercept term and that is why we see one more value than expected from a quadratic model with 1 input feature.

```python
# Building a model with no bias
pipe_no_bias = Pipeline([('quad_features', PolynomialFeatures(degree=2, include_bias=False)),
                ('quad_model', LinearRegression())])
pipe_no_bias.fit(auto[['horsepower']], auto['mpg'])
no_bias_mse= mean_squared_error(pipe_no_bias.predict(auto[['horsepower']]), auto['mpg'])
print(no_bias_mse)
```

```python
# cubic model
cubic_pipe = Pipeline([('quad_features', PolynomialFeatures(degree=3, include_bias=False)),
                ('quad_model', LinearRegression())])
cubic_pipe.fit(auto[['horsepower']], auto['mpg'])
cubic_mse= mean_squared_error(cubic_pipe.predict(auto[['horsepower']]), auto['mpg'])
print(cubic_mse)
```

```python
# Predicting the mpg for a vehicle with hp=200
cube_predict = cubic_pipe.predict([['200']])
print(cube_predict)
```
